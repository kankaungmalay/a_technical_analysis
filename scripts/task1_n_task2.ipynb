### Import Packages

```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
import warnings
warnings.filterwarnings('ignore')
plt.style.use('ggplot')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

## Task 1

```python
# absorbance readings of samples
df = pd.read_csv("../datasets/calibration.csv")
df
```

### Exploratory Data Analysis

```python
df.shape
```

```python
# row to column
df_converted = df.melt(id_vars=["Sample", "Dilution", "Well"], 
        var_name="Wavelength", 
        value_name="Value")
df_converted
```

```python
# checking % of missing values in columns
for col in df_converted.columns:
    pct_missing = np.mean(df_converted[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# checking data types
cat_cols = df_converted.select_dtypes(include = 'object')
num_cols =df_converted.select_dtypes(include = np.number)
print('Numerical variables are : ', num_cols.columns)
print('Categorical Columns are : ', cat_cols.columns)
```

### Visualisation

```python
# Scatter Plot
df_converted.plot(kind = 'scatter', x = 'Wavelength', y = 'Value')
plt.show()
```

```python
# Line Plot
sns.lineplot(x="Wavelength", y='Value', style="Dilution", data=df_converted)
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
# A facetgrid
graph = sns.FacetGrid(df_converted, col ="Sample",  hue ="Dilution")
# map the above form facetgrid with some attributes
graph.map(plt.scatter, "Wavelength", "Value", edgecolor ="w").add_legend()
# show the object
plt.show()
```
